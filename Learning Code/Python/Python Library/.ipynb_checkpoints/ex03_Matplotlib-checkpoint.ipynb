```python
# 라이브러리 로딩하기
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# line chart 그리기
y = [2,4,6,8]
plt.plot(y)
# 하나의 데이터를 넣어줄 경우 y 축으로 가고 x 축은 자동 부여 (0 ~)
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
# plt.plot(x축,y축)
plt.show()
```

```python
# Line Chart 그리기
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y,ls = '--', lw = 5, color = 'm', marker = 'o', mfc = 'y')

# 속성값
# 1. 라인의 스타일 설정가능 ls(LineStyle) --> 문자열로 기호를 적어주기!
# 2. 라인 두께 설정 가능 lw(Line Width) --> 그냥 숫자로 써주기~ 문자 XX
# 3. 라인 색 변경 가능 color -> marker와 line의 색 모두 변경
# 4. 값의 위치 표시하기 - marker
# 5. maker 색 지정하기 - mfc(Marker Face Color)

plt.grid()
# grid 속성
# 모눈종이 처럼 생긴 격자 무늬 -> 그래프를 보기가 쉬워진다

plt.show()
```

```python
# 연습문제
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]

plt.plot(x,y, ls='--', lw = 2, color = 'g', marker = 'o', mec = 'g')

plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, ls = '-.', lw = 5, color = '#C79FEF', marker = '$❤$', ms = 20, mfc = 'w',mec = '#FF0000', mew = 2)

# 그래프의 범위 지정 (xlim, ylim)
plt.xlim(-2, 10)
plt.ylim(-5, 10)
# plt.xlim(2,6)
# plt.ylim(3,7) --> 확대

# 그리드 : 격자무늬 -> 가시성이 높아진다.
plt.grid()

# ticks : 눈금표시
plt.xticks([0,1,2])
plt.yticks(y)

plt.show()
```

### 하나의 차트에 여러개의 plot 출력하기

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

# 시각화 하는 이유는  -> 한눈에 데이터를 보기 위함

plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')

# 범례 설정하기 -> 참고사항
# plot 해줄때 label 이름 적어주기!
plt.legend(loc='best')

# x축 이름 설정하기
plt.xlabel('x_lable')
# y축 이름 설정하기
plt.ylabel('y_lable')

# Chart 이름 설정하기
plt.title('차트')

plt.show()
```

### 요일별 교통사고 차트 그리기 실습

```python
# csv 불러오기!
data = pd.read_csv('Traffic_Accident_2017.csv', encoding='EUC-KR')
```

```python
# DataFrame의 정보를 알려주는 기능
data.info()
```

```python
# 상위 데이터만 확인하는 방법
data.head(n=3)     # 상위 5개의 데이터만 출력해서 보여준다!
```

```python
# data에서 우리가 필요한 '요일'만 가져오기
data['요일'].count()
```

```python
# 각각의 요일별로 얼마나 발생했는지?
data2 = data['요일'].value_counts()
data2
```

```python
# 요일 정렬하기 월 화 수 목 금 토 일
data3 = data2[['월','화','수','목','금','토','일']]
```

```python
# bar 그래프 그려주기!
x = data3.index
y = data3.values
plt.bar(x,y, width=0.3, color='m', edgecolor='black')

# x축 : 요일
plt.xlabel('요일')
# y축 : 교통사고 발생건수
plt.ylabel(' 교통사고 \n발생 건수', labelpad=40).set_rotation(0)
# 차트이름 : 요일별 교통사고 발생건수
plt.title('요일별 교통사고 발생 건수')

# 테이터의 차이를 구분할 수 있도록 -> 확대, 축소
plt.ylim(500, 620)

plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
# 모든 컬럼을 출력하는 방법
pd.set_option('display.max_columns',None)
data
```

```python
# 1. 사고유형_대분류에서 '차대차' 사건만 가져오기 --> 불리언 인덱싱
acc = data[data['사고유형_대분류'] == '차대차']
```

```python
# 총데이터에서 각각 얼마나 사고가 발생했는지
data['사고유형_대분류'].value_counts()
```

```python
# 우리가 필요한 사상자수와 발생지 시도 데이터만 가져오기
data1 = acc[['사상자수','발생지시도']]
```

```python
# 각각의 지역이 몇번 나왔는지만 세어주는 결과값 -> 우리가 원하는 데이터 아님
data1['발생지시도'].value_counts()
```

```python
# groupby()
# 같은 값을 하나로 묶어서 총계를 구해주고 싶을때 사용한다
result = acc[['사상자수','발생지시도']].groupby('발생지시도', as_index = False).sum()
# groupby()함수를 사용하면 기준 열이 자동으로 인덱스로 간다
# 만약 index로 사용하지 않고 싶다면? as_index = False
new_result = result['사상자수'].values
new_result
```

```python
x = result['발생지시도'].values
y = new_result

# 차트 사이즈 조절 ※ 차트 정의 전에 적어줘야한다!
plt.figure(figsize=(8,6))
plt.bar(x,y,width=0.7)

plt.xlabel('지역', size=15)
plt.ylabel('사상자수', size=15, labelpad=30).set_rotation(0)
plt.title('차 vs 차 교통사고의 사상자 수', size=25)

plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8)

plt.show()
```

```python
# 파이차트
# -> 범주형 데이터를 시각화 할때 많이 사용한다.
fruits = ['수박','복숭아','바나나','사과']
ratio = [18,24,45,13]
color = ['r','y','m','c']
explode = [0,0,0.1,0]
plt.pie(ratio, labels = fruits, colors = color, shadow = True, explode = explode)
plt.show()
```
