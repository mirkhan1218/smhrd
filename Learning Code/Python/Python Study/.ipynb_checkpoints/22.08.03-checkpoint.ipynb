### 리스트
- 파이썬의 자료구조 형태 중 하나
- 순서가 있는 수정 가능한 객체의 집합이다.
- 대괄호로 작성이 되어지며 리스트 내부의 값은 콤마로 구분된다.
- 추가, 수정, 삭제 가능하다.

```python
# 리스트에 값 추가 -> append(값) -> 맨뒤에 데이터가 추가된다!
list6 = [0,1,2,3,4]
list6.append(6)
list6.append(5)
print(list6)
```

```python
fruit = ['사과','포도',['수박','멜론'],'복숭아','딸기','오렌지']
a = fruit[0]
w= fruit[2][0]
o = fruit[5]

choice_list=[]

choice_list.append(a)
choice_list.append(w)
choice_list.append(o)

print(choice_list)
```

```python
# 리스트에 값 추가하기 -> insert(인덱스위치,값)
list7 = [0,1,2,3,4,5,6]
list7.insert(7,7)
print(list7)
```

```python
music_list=[['Butter','방탄'],['헤픈우연','헤이즈'],['라일락','아이유']]
loc = int(input('인덱스 위치 입력 >> '))
song = input('노래 제목 입력 >> ')
singer = input('가수 입력 >>')

temp = [song,singer]

music_list.insert(loc,temp)
print(music_list)
```

```python
# 리스트 삭제 -> del 키워드
list9 = [0,1,2,3,4,5]
del list9[4:]
print(list9)
```

```python
# 리스트 삭제 -> remove()
list10 = ['a','b','c','d','e']
list10.remove('c')
print(list10)
```

```python
# 리스트 값 정렬 -> 오름차순으로 정렬 -> sort()
list11 = [9,77,13,52,100,3]
list11.sort()
print(list11)
```

```python
# 역순으로 정렬 -> reverse()
list11 = [9,77,13,52,100,3]
list11.reverse()
print(list11)
```

```python
list11 = [9,77,13,52,100,3]
# list11.sort()
# list11.reverse()
list11.sort(reverse=True)
print(list11)
```

```python
# 리스트 위치 반환 함수 -> index()
list12 = ['a','b','c','d','e','f']
list12.index('d')
```

```python
# 마지막 값을 반환 후 리스트에서 제거 -> pop()
list12 = ['a','b','c','d','e','f']
print(list12.pop())
print(list12)
```

```python
list12 = ['a','b','c','d','e','f']
len(list12)
```

```python
list1 = ['딸기','바나나','수박','체리','포도']
word = input("검색할 문자를 입력하세요 >> ")

if word in list1:
    print(f'{word}는 리스트에 {list1.index(word) + 1}번째 인덱스에 들어있습니다.')
else:
    print(f'{word}는 리스트에 들어있지 않습니다.')
```

```python
s = "Hi, My name is Mihee"

input_s = input("검색할 문자를 입력하세요 : ")

if input_s in s:
    print(f'{input_s}는 {s.count(input_s)} 번 들어 있습니다')
else:
    print(f'{input_s}는 문자열에 들어있지 않습니다')    
```

```python
# while문을 사용해서 "파이썬 최고" 13번 출력하기
i = 0
while i < 13:
    print(f'파이썬 최고 {i + 1}')
    i += 1
```

```python
num = 1
while True:
    print(num)
    num += 1
    if num > 3:
        break
```

```python
while True:
    num1 = int(input("첫 번째 정수 입력 >> "))
    num2 = int(input("두 번째 정수 입력 >> "))
    
    if num1 == 0 and num2 == 0:
        print("프로그램이 종료되었습니다.")
        break
    
    print(f'두 정수의 합 : {num1 + num2}')
```

```python
# 랜덤 라이브러리
import random
```

```python
# 1~50 사이 숫자 랜덤하게 추출
random.randint(1,10)
```

```python
# Shift + Tab 함수 설명
num = random.randint(1,50)

while True:
    ans = int(input('숫자를 입력하세요 >> '))
    if num < ans:
        print(f'{ans}보다 작은수 입니다.')
    elif num > ans:
        print(f'{ans}보다 큰수 입니다.')
    else:
        break
print('정답을 맞추셨습니다')
```

### for 문


```python
say = "치킨 먹고 싶다"
for s in say:
    print(s)
```

```python
snack_tuple =('아우터','차카니','초코픽')

for snack in snack_tuple:
    print(snack)
```

```python
score_list = [90, 45, 70, 60, 55]

i = 0

for score in score_list:
    i += 1
    if score >= 60:
        print(f'{i}번 학생은 합격입니다')
    else:
        print(f'{i}번 학생은 불합격입니다')
```

#### range 함수
- range(시작, 종료(포함X), 증가)
- range(기본값 0, 종료, 기본값 1)

```python
for i in range(1, 10, 1):
    print(i, end = ' ')    # 기본값(디폴트값) = '\n'(개행)
```

```python
for i in range(6, 10):    # 6~9. 기본값 증가량 1
    print(i, end = ' ')
```

```python
for i in range(6):    # 0~5. 기본값 증가량 1
    print(i, end = ' ')
```

```python
for i in range(97, 76, -1):
    print(i, end = ' ')
```

```python
for i in range(23, 40, 1):
    print(i, end = ' ')
```

```python
a, b = 3, 20
print(a)
print(b)
```

```python
list1 = [[1,2],[3,4],[5,6]]

for i, j in list1:
    print(i)
    print(j)
    print('반복중')
```
