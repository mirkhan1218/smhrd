#### 2. 리뷰 길이 최대 값, 리뷰 내용

```python
# 리뷰 길이
# 리스트
max_len = 0
max_index = 0
for i in range (len(my)):
    if max_len < len(my[i][0]):
        max_index = i
        max_len = len(my[i][0])
print(f'({max_len}) 내용 : {my[max_index][0]}')
```

```python
# 리스트
max_len = len(my[0][0])
max_content = ''
for re in my:    # for 문 안에서 len(re[0]) 는 리뷰 길이
    r_len = len(re[0])   # r_len - 현재 리뷰 길이
    if max_len < r_len:
        max_len = r_len
        max_content = re[0]
print('리뷰 최대 길이는 %d입니다.'%max_len)
print(max_content)
```

```python
min_len = len(my[0][0])
min_index = 0
for i in range (len(my)):
    if min_len > len(my[i][0]):
        min_index = i
        min_len = len(my[i][0])
print(f'({min_len}) 내용 : {my[min_index][0]}')
```

```python
re_add = 0
for re in my:
    re_add += len(re[0])
re_avg = re_add / len(my)
print(f'리뷰의 평균 길이는 {re_avg}입니다.')
```

```python
re_pl = 0
re_mi = 0

for re in my:
    re_re = re[1]
    if re_re == 0:
        re_mi += 1
    else:
        re_pl += 1

print(f'긍정 {re_pl}개')
print(f'부정 {re_mi}개')
```

```python
re_list = []
for re in my:
    re_list.append(re[1])    # re[1] -> 0/1
    
n_cnt = re_list.count(0)
p_cnt = re_list.count(1)

print(f'긍정 {n_cnt}개')
print(f'부정 {p_cnt}개')
```

```python
count = 0

for re in my:
    if 'ㅋ' in re[0]:
        count += 1
        
print(f'ㅋ가 포함된 리뷰는 {count}개 입니다.')
```

```python
literal = input('검색할 문자 입력 >> ')

count = 0

for re in my:
    if literal in re[0]:
        count += 1
        
print(f'{literal}가 포함된 리뷰는 {count}개 입니다.')
```
