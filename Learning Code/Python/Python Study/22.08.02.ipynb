## 삼항연산자
- a if 조건식 else b

```python
score = 80
# 점수가 60점 이상이면 합격 미만이면 불학격
print("합격") if score >= 60 else print("불합격")
```

```python
score = 50
# 점수가 60점 이상이면 합격 미만이면 불학격
print("합격") if score >= 60 else print("불합격")
```

```python
a = int(input("정수입력 >> "))
b = int(input("정수입력 >> "))
print('a : {}'.format(a)) if a > b else print('b : {}'.format(b))
```

```python
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))

sum = num1 - num2 if num1 > num2 else num2 - num1

print(f'두 수의 차 : {sum}')
```

```python
score = int(input("점수 입력 >> "))

if score >= 90:
    print(f'{score}점은 A학점 입니다')
elif score >= 80:
    print(f'{score}점은 B학점 입니다')
elif score >= 70:
    print(f'{score}점은 C학점 입니다')
elif score > 60:
    print(f'{score}점은 D학점 입니다')
else:
    print(f'{score}점은 F학점 입니다')
```

```python
# 키가 120cm 이상일때 놀이기구 탑승 가능
# 나이에 따라 다른 요금 -> 12세 미만 500원 12세이상 18세이하 1000원 18세 초과 3000원
# 사진을 찍는 여부에 따라 요금 달라짐 (사진을 찍는다면 +300원)
height = int(input("키 몇?"))
result = 0

if height >= 120:
    print("놀이기구 탑승 가능")
    age = int((input("나이 몇?")))
    if age > 18:
        result = 3000
    elif age >= 12:
        result = 1000
    else:
        result = 500
        
    picture = input("사진 찍으실?")
    if picture == "y":
        result += 300
        
    print(f'총 비용은 {result} 입니다.')

else:
    print("놀이기구 탑승 불가")
```
