#### Selenium 모듈
- 웹 브라우저를 자동으로 제어하기 위한 라이브러리
- 브라우저 띄우는것부터 제어 -> requests가 필요가 없음

#### find_element()
- By.ID : 아이디 선택자 이름

- By.CSS_SELECTOR : CSS선택자 이름
- By.TAG_NAME : 태그를 이용해서 선택

```python
# 현재 페이지에서 5번 스크롤링, 한번 스크롤링 할떄 1초씩 쉬기
body = driver.find_element(By.TAG_NAME,'body')

for i in range(5):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
```

```python
# res.content or res.text = 정적인 페이지
# driver.page_soutce = 동적인 페이지
soup = bs(driver.page_source,'lxml')
```

```python
# 영상제목, 영상주소, 조회수
title = soup.find_all('a',id='video-title')
len(title)
```

```python
nums = []
titles = []
video_urls = []

length = soup.find_all('a',id='video-title')

for i in range(len(length)):
    title = soup.find_all('a', id='video-title')[i].text.strip()
    video_url = soup.find_all('a', id='video-title')[i]['href']
    
    nums.append(i+1)
    titles.append(title)
    video_urls.append(video_url)
    
    print("영상제목 : " + title)
    print("영상주소 : " + url_main + video_url)
    print("-"*100)
    
len(nums), len(titles), len(video_urls)
```

```python
# youtube.csv 저장하기
df = {'번호':nums, '영상제목':titles, '영상주소':video_urls}
youtube = pd.DataFrame(df)
youtube.set_index('번호', inplace=True)
youtube.head()

youtube.to_csv('youtube.csv', encoding='utf-8')
```

#### 네이버 뉴스 제목, 내용 가져오기

```python
url_main = "https://n.news.naver.com"
url_sub = "/mnews/article/055/0000994481?sid=105"

#1. 크롬창 열기
driver = wb.Chrome()
# 2. 네이버 > 뉴스 접속
driver.get(url_main + url_sub)

soup = bs(driver.page_source,'lxml')

head_line = soup.find('h2', 'media_end_head_headline').text
content = soup.find('div', id='dic_area').text.strip().replace("\n","")

print("제목 : " + head_line)
print("내용 : " + content)
```

#### 네이버 책에서 저자하고 책제목 가져오기

```python
url = "https://book.naver.com"


#1. 크롬창 열기
driver = wb.Chrome()
# 2. 네이버 > 책 접속
driver.get(url)

soup = bs(driver.page_source,'lxml')

# 네이버책 제목, 저자
titles = soup.select('.rank_view+dl>dt+dt')
authors = soup.select('.rank_view+dl>dd')

len(titles), len(authors)
```

```python
for i in range(len(titles)):
    print("제목 : " + titles[i].text)
    print("저자 : " + authors[i].text)
    print("-"*100)
```

```python
# 상품 5개에 대한 정보를 수집
ranks = []
title_list = []
price_list = []

url = "http://corners.gmarket.co.kr/Bestsellers"

driver = wb.Chrome()
driver.get(url)

for i in range(200):
    imgs = driver.find_elements(By.CLASS_NAME,'lazy')

    # i번째 클릭
    imgs[i].click()
    
    # 1초 쉬기
    time.sleep(1)
    
    # 1. 상품명 가격 가져오기
    title=driver.find_element(By.CLASS_NAME,'itemtit').text
    price = driver.find_element(By.CLASS_NAME,'price_real').text
    
    # 2. 리스트에 추가하기
    ranks.append(i+1)
    title_list.append(title)
    price_list.append(price)
    
    # 3. 상품명, 가격 출력하기
    print("상품명 : " + title)
    print("가격 : " + price)
    print("-" * 100)
    
    # 4. 뒤로가기
    driver.back()
    
    # 5. 1초 쉬기
    time.sleep(1)
```

```python
# 6. gmarket_best.csv로 저장하기
df = pd.DataFrame({'상품명':title_list, '가격':price_list})
df.head()
```

```python
df.to_csv('gmarket_best.csv', encoding='utf-8')
```
