#### string, text 차이
- string 태그 하위에 문자열을 객체화, 문자열이 없으면 None 반환
- text 태그 하위에 있는 자식 텍스트를 문자열로 반환해줌. 문자열이 없으면 출력이 없음

```python
# 분석대상
html = """
<html>
<body>
  <h1>글자크기</h1>
  <h2>글자크기2</h2>
  <p>원하는 글자 추출</p>
  <p>크롤링 수업시간 </p>
  <p></p>
  <ul>
      <li><a href="http://www.daum.net">daum</a></li>
      <li><a href="http://www.naver.com">naver</a></li>
  </ul>
</body>
</html>  """
```

```python
soup = bs(html,'lxml')
```

```python
data = soup.find("h1")
```

```python
data.string
```

- id : 문서안에서 한번만 사용
- class : 공통적인 내용을 적용을 할때 같은 이름으로 중복 가능

```python
html = """
<html>
<body>
  <h1 id="title">글자크기</h1>
  <p class="cstyle">원하는 글자 추출</p>
  <p id="body" align="center">크롤링 수업시간 </p>
</body>
</html>
"""
```

```python
soup = bs(html,'lxml')
```

```python
# h1태그에 id가 'title'를 지정해서 필요한 데이터 추출하기
data = soup.find('h1', id='title').text
```

```python
print(data)
```

#### css 선택자 사용하기
- soup.select_one("선택자") : 선택자 요소 하나를 추출
- soup.select("선택자") : 선택자 요소 여러개를 리스트로 추출

```python
html = """
<html>
<body id="me">
<div id="smart">
      <h1 id="title">스마트 미디어 도서</h1>
      <ul class="items">
          <li>파이썬입문</li>
          <li>머신러닝</li>
          <li>딥러닝입문</li>
      </ul>
 </div>
</body>
</html>
"""
```

```python
soup = bs(html,'lxml')
```

```python
# id는 #id값, class는 .class값으로 표기
data = soup.select_one("div#smart>h1").text
```

```python
data
```

```python
data = soup.select("div#smart>ul.items>li")
data
```

```python
for i in data:
    print(i.text)
```

#### 네이버 검색 페이지 가져오기

```python
keyword = "쿠키런 킹덤"
url = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=" + keyword
res = req.get(url)

data = bs(res.content,'lxml')

cookie = data.select_one('div.title_area.type_keep._title_area > h2 > span > strong')
cookie.text
```

```python
cookie_names = data.select('div.title_box>strong>a')
cookie_names
```

```python
cookie_name_list=[]
for i in cookie_names:
    cookie_name_list.append(i.text)
    print(i.text)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(cookie_name_list,columns=['쿠키이름'])
df
```

```python
df.to_csv('cookie_names.csv', encoding="UTF-8")
```
