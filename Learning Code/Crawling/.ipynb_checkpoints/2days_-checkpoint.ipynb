#### 1. 위키문헌 윤동주 작품 리스트 가져오기

```python
res = req.get("https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC")
res
```

#### .text와 .content 차이
- .text : HTTP Request를 보낸 URL에서 읽어올수 있는 내용을 가져올때 사용,
    requests를 자동으로 인코딩을 추출하고 디코딩시킨값으로 한번 가공된 값.
    원본 파일을 볼수없음
- .content : 디코드 되지않은 차이, 원본 파일을 볼수있음.

```python
soup = bs(res.content,"html.parser")
```

```python
# > : 자식태그, 공백 : 후손태그
li = soup.select("#mw-content-text > div.mw-parser-output > ul > li a")
```

```python
# 반복문으로 작품 이름 출력 : - 작품이름
for i in li:
    name = i.text
    print("-",name)
```

```python
import pandas as pd
```

```python
# 리스트에 작품 넣기
name_list=[]
for i in li:
    name_list.append(i.text)
```

```python
df = pd.DataFrame(name_list, columns=["작품"])
```

```python
df.to_csv("work_list.csv", index=False, encoding="utf-8")
```

#### 2. yes24 베스트셀러 작품가져오기

```python
res = req.get("http://www.yes24.com/24/Category/BestSeller").text
```

```python
soup = bs(res,'lxml')
```

```python
# 책 설명 출력
books = soup.select("#bestList > ol > li > p.copy > a")
```

```python
for i in books:
    title = i.text
    print(title)
```

```python
# 책 가격 가져오기
prices = soup.select("#bestList > ol > li > p.price > strong")
```

```python
for i in prices:
    price = i.text
    print(price)
```

```python
# 책 제목 가져오기
books = soup.select("#bestList > ol > li")
```

```python
for i in books:
    title = i.select('p > a')[2].text
    price = i.select('p.price > strong')[0].text
    print(title, " : ", price)
```

#### select() 정리
- tag 이름 : select("h1")
- tag+class이름 : select("h1.클래스이름")
- class이름 : select(".클래스이름")
- id이름 : select("#id이름")
- 상하위 태그 선택 가능
    - 공백 (후손태그) : select("html head title")
    - ">" (자식태그) : select("head>title")

#### 4. 네이버쇼핑(노트북 모델명과 가격출력)

```python
res = req.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EB%85%B8%ED%8A%B8%EB%B6%81")
soup = bs(res.content, 'lxml')
```

```python
# find_all() 사용 모델명과 가격
models = soup.find_all('a','title elss')
prices = soup.find_all('div','price')
len(models), len(prices)
```

```python
for i in range(len(models)):
    print(str(i+1) + " " + models[i].text + " / " + prices[i].text)
```
